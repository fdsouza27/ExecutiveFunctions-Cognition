```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df_CP=pd.read_csv('Features_ComplexPie.csv')
df_SP=pd.read_csv('Features_SimplePie.csv')
```

```python
df_CP
```

### Complex Pie Vs Simple Pie

#### 1) Total Actions

```python
t_stat, p_value = ttest_ind(df_CP['Total Actions'], df_SP['Total Actions'], equal_var=False)

plt.figure(figsize=(7, 5))
plt.boxplot([df_CP['Total Actions'], df_SP['Total Actions']], labels=['Complex Pie', 'Simple Pie'], widths = 0.2)
plt.title('Total Actions for Complex and Simple Pie')
plt.ylabel('Action Count')
plt.text(0.95, 0.95, f't-value = {t_stat:.2f}\nP-value = {p_value:.2e}', 
         transform=plt.gca().transAxes,
         fontsize=9, verticalalignment='top', horizontalalignment='right', 
         bbox=dict(facecolor='white', alpha=0.2))

directory = 'C:/Users/FANNY MARIA D SOUZA/Capstone/'
folder_path = os.path.join(directory, 'Plots/Boxplot_Pie')


file_path = os.path.join(folder_path, '1 Pie_TotalActions.png')
plt.savefig(file_path, dpi=300, bbox_inches='tight')
plt.show()
```
