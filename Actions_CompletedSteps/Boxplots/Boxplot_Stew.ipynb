```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df_CS=pd.read_csv('Features_ComplexStew.csv')
df_SS=pd.read_csv('Features_SimpleStew.csv')
```

```python
df_CS
```

```python
df_SS
```

#### 1) Total Actions

```python
t_stat, p_value = ttest_ind(df_CS['Total Actions'], df_SS['Total Actions'], equal_var=False)

plt.figure(figsize=(7, 5))
plt.boxplot([df_CS['Total Actions'], df_SS['Total Actions']], labels=['Complex Stew', 'Simple Stew'])
plt.title('Total Actions for Complex and Simple Stew')
plt.ylabel('Action Count')
plt.text(0.99, 0.95, f't-value = {t_stat:.2f}\nP-value = {p_value:.2e}', 
         transform=plt.gca().transAxes,
         fontsize=8, verticalalignment='top', horizontalalignment='right', 
         bbox=dict(facecolor='white', alpha=0.2))
directory = 'C:/Users/FANNY MARIA D SOUZA/Capstone/'
folder_path = os.path.join(directory, 'Plots/Boxplot_Stew')


file_path = os.path.join(folder_path, '1 Stew_TotalActions.png')
plt.savefig(file_path, dpi=300, bbox_inches='tight')
plt.show()
```
