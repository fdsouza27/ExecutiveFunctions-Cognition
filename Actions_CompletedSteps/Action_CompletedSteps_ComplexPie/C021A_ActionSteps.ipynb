```python
# Sort the DataFrame by 'Session Time'
df_sorted = df.sort_values(by='Session Time').reset_index(drop=True)

# Find indices where steps are completed
completed_indices = df_sorted.index[df_sorted['Completed Step'].notna()].tolist()

# Initialize the actions_per_step, time_per_step, and missed_interactions_per_step lists
actions_per_step = []
time_per_step = []
missed_interactions_per_step = []

# Calculate actions performed, time spent, and missed interactions for each completed step
previous_index = 0
for current_index in completed_indices:
    if pd.notna(df_sorted.loc[current_index, 'Action Type']):
        actions_count = current_index - previous_index + 1  # Include the current action if it's not null
    else:
        actions_count = current_index - previous_index  # Do not include the current action if it's null
    actions_per_step.append(actions_count)

    # Calculate time spent on this step
    time_spent = df_sorted.loc[current_index, 'Session Time'] - df_sorted.loc[previous_index, 'Session Time']
    time_per_step.append(time_spent)
    
    # Calculate missed interactions
    missed_interactions_count = df_sorted.loc[previous_index:current_index, 'Missed Interactions'].sum()
    missed_interactions_per_step.append(missed_interactions_count)
    
    previous_index = current_index + 1  # Move to the next action after the completed step

# Append the actions, time, and missed interactions for the period after the last completed step
actions_per_step.append(len(df_sorted) - previous_index)
time_per_step.append(df_sorted['Session Time'].iloc[-1] - df_sorted.loc[previous_index, 'Session Time'])
missed_interactions_per_step.append(df_sorted.loc[previous_index:, 'Missed Interactions'].sum())

# Create a dataframe to display the results
result_df = pd.DataFrame({
    'Completed Step': df_sorted.loc[completed_indices, 'Completed Step'].tolist() + ['Post last step'],
    'Actions Count': actions_per_step,
    'Time Spent': time_per_step,
    'Missed Interactions': missed_interactions_per_step
})

result_df
```

```python
# Calculate the total values
total_actions = result_df['Actions Count'].sum()
total_time_spent = result_df['Time Spent'].sum()
total_missed_interactions = result_df['Missed Interactions'].sum()

# Create a new DataFrame with these totals
totals_df = pd.DataFrame({
    'Participant' : 'C021',
    'Total Actions': [total_actions],
    'Total Time Spent': [total_time_spent],
    'Total Missed Interactions': [total_missed_interactions]
})

totals_df
```

```python
total_actions = result_df['Actions Count'].sum()
total_actions
```

```python
df_grouped = result_df.groupby("Completed Step").agg({
    "Actions Count": "sum",
    "Time Spent": "sum"
}).reset_index()

# Plotting
fig, ax1 = plt.subplots(figsize=(20, 6))

color = 'tab:blue'
ax1.set_xlabel('Completed Step')
ax1.set_ylabel('Actions Count', color=color)
ax1.bar(df_grouped['Completed Step'], df_grouped['Actions Count'], color=color, alpha=0.6, label='Actions Count')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(0, 250)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Time Spent', color=color)
ax2.plot(df_grouped['Completed Step'], df_grouped['Time Spent'], color=color, marker='o', label='Time Spent')
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, 250)

fig.tight_layout()
plt.xticks(rotation=45, ha='right')
plt.title('Visualization of Actions Count and Time Spent by Completed Step')
plt.show()
```

```python
df_sorted = df.sort_values(by='Session Time').reset_index(drop=True)
completed_indices = df_sorted.index[df_sorted['Completed Step'].notna()].tolist()

# Initialize a dictionary to store the count of interactions per object under each step
interactions_per_step = {}


previous_index = 0

for current_index in completed_indices:
    step_name = df_sorted.loc[current_index, 'Completed Step']
    step_data = df_sorted.iloc[previous_index:current_index+1]
    object_counts = step_data['Interacted Object'].value_counts().to_dict()
    
    if step_name in interactions_per_step:
        # If the step is already in the dictionary, add the counts to existing values
        for obj, count in object_counts.items():
            if obj in interactions_per_step[step_name]:
                interactions_per_step[step_name][obj] += count
            else:
                interactions_per_step[step_name][obj] = count
    else:
        interactions_per_step[step_name] = object_counts
    
    previous_index = current_index + 1

interactions_df = pd.DataFrame.from_dict(interactions_per_step, orient='index').fillna(0)
interactions_df = interactions_df.rename(columns=lambda x: x.split('_')[0])
interactions_df = interactions_df.T.groupby(interactions_df.columns).sum().T
interactions_df = interactions_df.astype(int)
interactions_df['Interacted Objects']=interactions_df.sum(axis=1)

interactions_df
```
