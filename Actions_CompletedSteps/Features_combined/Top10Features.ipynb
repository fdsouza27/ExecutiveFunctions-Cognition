```python
import pandas as pd
from scipy.stats import spearmanr
```

```python
CP=pd.read_csv('CombinedFeatures_ComplexPie.csv')
```

```python
CP
```

```python
exclude_columns = ['Participant']
target_column = 'Block Design Raw Score'

columns_to_analyze = [col for col in CP.columns if col not in exclude_columns and col != target_column]

results = []
for column in columns_to_analyze:
    if CP[column].dtype in ['int64', 'float64']:  
        correlation, p_value = spearmanr(CP[column], CP[target_column], nan_policy='omit')
        results.append((column, correlation, p_value))

results_df = pd.DataFrame(results, columns=['Feature', 'Spearman Correlation', 'P-Value'])
results_df = results_df.sort_values(by='Spearman Correlation', key=abs, ascending=False)
top_10_features = results_df.head(10)
top_10_features
```
