```python
import pandas as pd
```

```python
action_CP=pd.read_csv('Features_ComplexPie_Action.csv')
```

```python
action_CP
```

```python
movement_CP=pd.read_csv('Features_ComplexPie_Movement.csv')
```

```python
movement_CP
```

```python
Combined_ComplexPie=pd.merge(action_CP, movement_CP, on='Participant', how='inner')
Combined_ComplexPie
```

```python
Combined_ComplexPie.to_csv('CombinedFeatures_ComplexPie.csv')
```
