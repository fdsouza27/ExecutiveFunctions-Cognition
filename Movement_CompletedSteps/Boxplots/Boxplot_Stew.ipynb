```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
df_CS=pd.read_csv('Features_ComplexStew_Movement.csv')
df_SS=pd.read_csv('Features_SimpleStew_Movement.csv')
```

#### Total Distance

```python
t_stat, p_value = ttest_ind(df_CS['Chest Total Distance'], df_SS['Chest Total Distance'], equal_var=False)

plt.figure(figsize=(7, 5))
plt.boxplot([df_CS['Chest Total Distance'], df_SS['Chest Total Distance']], labels=['Complex Stew', 'Simple Stew'])
plt.title('Chest Total Distance for Complex and Simple Stew')
plt.ylabel('Distance')
plt.text(0.99, 0.95, f't-value = {t_stat:.2f}\nP-value = {p_value:.2e}', 
         transform=plt.gca().transAxes,
         fontsize=8, verticalalignment='top', horizontalalignment='right', 
         bbox=dict(facecolor='white', alpha=0.2))
directory = 'C:/Users/FANNY MARIA D SOUZA/Capstone_Movement/'
folder_path = os.path.join(directory, 'Plots_Movement/Boxplot_Stew')


file_path = os.path.join(folder_path, '1 Stew_ChestTotalDistance.png')
plt.savefig(file_path, dpi=300, bbox_inches='tight')
plt.show()
```
