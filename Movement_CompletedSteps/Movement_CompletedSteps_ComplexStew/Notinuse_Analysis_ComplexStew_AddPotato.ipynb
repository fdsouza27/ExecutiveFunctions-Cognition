```python
final_df=pd.read_csv('ComplexStew_MS_PerStep.csv')
```

```python
filtered_potato = final_df[final_df['Step'].str.strip() == 'Add Potato']
filtered_potato = filtered_potato[~filtered_potato['Participant'].isin(['C010', 'C022', 'C030'])]
filtered_potato
#C029 is not included
```

```python
from sklearn.impute import KNNImputer
import numpy as np

filtered_potato = filtered_potato.copy()
imputer = KNNImputer(n_neighbors=1)

numerical_cols = filtered_potato.select_dtypes(include=['float64', 'int64']).columns
filtered_potato[numerical_cols] = filtered_potato[numerical_cols].replace(0, np.nan)

filtered_potato[numerical_cols] = imputer.fit_transform(filtered_potato[numerical_cols])
filtered_potato
```

```python
new_column_names = {
    'Chest Distance': 'Chest Distance - Add Potato',
    'HeadCamera Distance': 'HeadCamera Distance - Add Potato',
    'LeftElbow Distance': 'LeftElbow Distance - Add Potato',
    'LeftShoulder Distance': 'LeftShoulder Distance - Add Potato',
    'MotionControllerLeft Distance': 'MotionControllerLeft Distance - Add Potato',
    'MotionControllerRight Distance': 'MotionControllerRight Distance - Add Potato',
    'RightElbow Distance': 'RightElbow Distance - Add Potato',
    'RightShoulder Distance': 'RightShoulder Distance - Add Potato'
}

# Rename columns
filtered_potato = filtered_potato.rename(columns=new_column_names)
filtered_potato
```

```python
tracker_columns = ['Chest Distance - Add Potato', 'HeadCamera Distance - Add Potato', 'LeftElbow Distance - Add Potato', 'LeftShoulder Distance - Add Potato',
                   'RightElbow Distance - Add Potato', 'RightShoulder Distance - Add Potato', 'MotionControllerLeft Distance - Add Potato',
                   'MotionControllerRight Distance - Add Potato']
block_design_col = 'Block Design Raw Score'

# Function to calculate and display correlation for each tracker
def calculate_correlations(filtered_potato, tracker_columns, block_design_col):
    correlations = []
    for tracker_col in tracker_columns:
        # Calculate Spearman correlation and p-value without any filtering
        correlation, p_value = spearmanr(filtered_potato[block_design_col], filtered_potato[tracker_col])
        correlations.append({
            'Tracker': tracker_col,
            'Correlation': correlation,
            'P-value': p_value
        })
    corr_df = pd.DataFrame(correlations)
    return corr_df

# Calculate and display correlations
correlation_df = calculate_correlations(filtered_potato, tracker_columns, block_design_col)
correlation_df
```
