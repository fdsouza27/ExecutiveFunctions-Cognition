#### Step - Add bottom crust

```python
filtered_bottomcrust = final_df[final_df['Step'].str.strip() == 'Add bottom crust']
filtered_bottomcrust
```

```python
from sklearn.impute import KNNImputer
import numpy as np

filtered_bottomcrust = filtered_bottomcrust.copy()
imputer = KNNImputer(n_neighbors=1)

numerical_cols = filtered_bottomcrust.select_dtypes(include=['float64', 'int64']).columns
filtered_bottomcrust[numerical_cols] = filtered_bottomcrust[numerical_cols].replace(0, np.nan)

filtered_bottomcrust[numerical_cols] = imputer.fit_transform(filtered_bottomcrust[numerical_cols])
filtered_bottomcrust
```

```python
new_column_names = {
    'Chest Distance': 'Chest Distance - Add BottomCrust',
    'HeadCamera Distance': 'HeadCamera Distance - Add BottomCrust',
    'LeftElbow Distance': 'LeftElbow Distance - Add BottomCrust',
    'LeftShoulder Distance': 'LeftShoulder Distance - Add BottomCrust',
    'MotionControllerLeft Distance': 'MotionControllerLeft Distance - Add BottomCrust',
    'MotionControllerRight Distance': 'MotionControllerRight Distance - Add BottomCrust',
    'RightElbow Distance': 'RightElbow Distance - Add BottomCrust',
    'RightShoulder Distance': 'RightShoulder Distance - Add BottomCrust'
}

# Rename columns
filtered_bottomcrust = filtered_bottomcrust.rename(columns=new_column_names)
filtered_bottomcrust
```

```python
tracker_columns = ['Chest Distance - Add BottomCrust', 'HeadCamera Distance - Add BottomCrust', 'LeftElbow Distance - Add BottomCrust', 'LeftShoulder Distance - Add BottomCrust',
                   'RightElbow Distance - Add BottomCrust', 'RightShoulder Distance - Add BottomCrust', 'MotionControllerLeft Distance - Add BottomCrust',
                   'MotionControllerRight Distance - Add BottomCrust']
block_design_col = 'Block Design Raw Score'

# Function to calculate and display correlation for each tracker
def calculate_correlations(filtered_bottomcrust, tracker_columns, block_design_col):
    correlations = []
    for tracker_col in tracker_columns:
        # Calculate Spearman correlation and p-value without any filtering
        correlation, p_value = spearmanr(filtered_bottomcrust[block_design_col], filtered_bottomcrust[tracker_col])
        correlations.append({
            'Tracker': tracker_col,
            'Correlation': correlation,
            'P-value': p_value
        })
    corr_df = pd.DataFrame(correlations)
    return corr_df

# Calculate and display correlations
correlation_df = calculate_correlations(filtered_bottomcrust, tracker_columns, block_design_col)
correlation_df
```
